```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def get_delta(pred,targets):
    delta = pred-targets
    return delta
```

```python
def get_weighted_delta(m_inputs,m_delta):
    weighted_delta = []
    for i in range(len(m_delta)):
        temp = m_delta[i]*m_inputs
        weighted_delta.append(temp)
    return np.array(weighted_delta)
```

```python
'''
[ 1 2 3 4] -> node 1
[ 2 4 5 3] -> node 2
'''
def update_weights(weights,weighted_delta):
    alpha = 0.001
    for i in range(weighted_delta.shape[0]):
        for j in range(weights.shape[1]):
            weights[i][j] -= alpha*weighted_delta[i][j]
    return weights   
```

```python
def neural_network(inputs,weight,target):
    layer_1  = inputs.dot(weight.T)
    delta = get_delta(layer_1,target)
    weighted_delta = get_weighted_delta(inputs,delta)
    new_weights = update_weights(weight,weighted_delta)
    
    print("prediction: ",layer_1)
    print("Delta: ",delta)
    print("Weighted Delta: ",weighted_delta)
    print("new weight",new_weights)
    
    return np.array(new_weights)
```

```python
data = pd.read_csv('cleaned_data.csv')
data.drop(['Unnamed: 0'],inplace=True,axis=1)
```

```python
targets = np.zeros((7000,10))
labels = np.array(data['label'])
for i in range(len(labels)):
    targets[i][labels[i]]=1
inputs = np.array(data.drop(['label'],axis=1),dtype=np.float32)
weights = np.zeros((10,784))
weights.shape
```

```python
def trainer(m_inputs,m_targets,m_weights):
    for i in range(500000):
        print("*********Iteration {} *********".format(str(i)))
        m_weights = neural_network(m_inputs[i%7000],m_weights,m_targets[i%7000])
    print('*********** Training Complete *********')
    return m_weights
```

```python
w = trainer(inputs,targets,weights)
```
